````python
def generate_color_palette(num_colors: int) -> List[Tuple[int, int, int]]:
    """
    Generates a color palette with a specified number of distinct colors.
    
    Parameters:
    - num_colors (int): Number of distinct colors to generate.
    
    Returns:
    - list of RGB tuples representing colors.
    """
    # could hardcode a big array of colours and then split it, but if something silly like 100 colours is chosen that would break it
    # maybe split the colourwheel up by 8 or so, alternate around it and then add say 20 to r,g or b to get the others, 
    # maybe an exception for black and white, also keeping the colours bright.
    def valid_contrast(colour):
        # Relative luminance https://www.w3.org/TR/WCAG20/#relativeluminancedef
        rel_lum = 0.2126 * colour[0] + 0.7152 * colour[1] + 0.0722 * colour[2] 
        white_rel_lum = 0.2126 * 255 + 0.7152 * 255 + 0.0722 * 255

        # contrast ratio https://www.w3.org/TR/WCAG20/#contrast-ratiodef
        con_ratio = (white_rel_lum + 0.05) / (rel_lum + 0.05)

        # minimum contrast ratio https://www.w3.org/TR/WCAG20/#visual-audio-contrast-contrast
        if con_ratio >= 4.5:
            return True

        return False
            
    r = 0
    g = 0
    b = 0
    interval = 255 // num_colors*3
    palette = []
    for i in range(num_colors):
        rod_choice = i%3
        match rod_choice:
            case 0:
                r+=interval
            case 1:
                g+=interval
            case 2:
                b+=interval

        # check its distinct based off of WCAG
        colour = (r,g,b)
        j = 0
        increment = 20
        limit = 255//increment
        while not valid_contrast(colour):
            r += increment
            colour = (r,g,b) 
            if j == limit: # prevent endless loop
                break 
            j+=1

        safe_colour = (colour[0]%256, colour[1]%256, colour[2]%256)
        palette.append(safe_colour)
        
    return palette

def scale_image(image: np.ndarray, num_pixels: int) -> np.ndarray:
    """
    Scales an image to the desired size.
    
    Parameters:
    - image (np.ndarray): The input image.
    - num_pixels (int): Desired size for width and height.
    
    Returns:
    - np.ndarray: Rescaled image.
    """
    # cv2 resize wrapper
    return cv2.resize(image, (num_pixels,num_pixels)) # not really scaling but it does say desired size for w & h

def check_overlap(center: Tuple[int, int], size: int, existing_shapes: List[dict]) -> bool:
    """
    Checks if a new shape overlaps with existing shapes.
    
    Parameters:
    - center (Tuple[int, int]): Center of the new shape.
    - size (int): Size of the new shape.
    - existing_shapes (List[dict]): List of existing shapes.
    
    Returns:
    - bool: True if there is an overlap, False otherwise.
    """
    # treats all shapes as rects, will be an overestimate for circles and triangles but no shape data available
    
    # aabb collision
    def to_aabb(center,size):
        min_x = center[0]-size
        min_y = center[1]-size
        max_x = center[0]+size
        max_y = center[1]+size

        return (min_x,min_y,max_x,max_y)

    def aabb_collision(box1,box2):
        return (box1[0] <= box2[2] and box1[2] >= box2[0]) and (box1[1] <= box2[3] and box1[3] >= box2[1])

    new_aabb = to_aabb(center,size)

    for shape in existing_shapes:
        existing_aabb = to_aabb(shape['center'], shape['size'])

        if aabb_collision(new_aabb, existing_aabb):
            return True

    return False
        
        

def draw_circle(image: np.ndarray, center: Tuple[int, int], size: int, color: Tuple[int, int, int]) -> float:
    """
    Draws a circle on an image.
    
    Returns:
    - float: Area of the circle.
    """
    cv2.circle(image, center, size, color, -1)
    return np.pi * np.square(size)

def draw_square(image: np.ndarray, center: Tuple[int, int], size: int, color: Tuple[int, int, int]) -> int:
    """
    Draws a square on an image.
    
    Returns:
    - int: Area of the square.
    """
    blc = (center[0]-size, center[1]-size)
    trc = (center[0]+size, center[1]+size)
    cv2.rectangle(image, blc, trc, color, -1)
    return int(np.square(2*size)) # size is half length

    return int(np.square(2*size)) # size is half length

def draw_triangle(image: np.ndarray, center: Tuple[int, int], size: int, color: Tuple[int, int, int]) -> float:
    """
    Draws a triangle on an image.
    
    Returns:
    - float: Area of the triangle.
    """
    rng = np.random.default_rng()

    top = (center[0], center[1]+size)

    def rotate_point(point, origin, rotation):
        px = point[0] - origin[0]
        py = point[1] - origin[1]
        
        rx = px * np.cos(rotation) - py * np.sin(rotation)
        ry = px * np.sin(rotation) + py * np.cos(rotation)

        tx = int(rx+origin[0])
        ty = int(ry+origin[1])
        
        return (tx,ty)
    
    deg120 = np.deg2rad(120)
    deg240 = np.deg2rad(240)
    p1 = rotate_point(top,center,deg120)
    p2 = rotate_point(top,center,deg240)

    # random rotation, as its about center shouldn't affect collision code
    degrees = rng.integers(0,360)
    radians = np.deg2rad(degrees)
    p3 = rotate_point(top, center, radians)
    p2 = rotate_point(p2, center, radians)
    p1 = rotate_point(p1, center, radians)
    triangle = np.array([[p1,p2,p3]])

    cv2.fillPoly(image,triangle,color)

    return np.sqrt(3)/4 * np.square(size*2)

def generate_image(img_size: int, num_pixels: int, color_palette: List[Tuple[int, int, int]], max_shapes: int = 10) -> Tuple[np.ndarray, List[dict]]:
    """
    Generates an image with at least one circle, one square, and one triangle.
    Uses helper functions for shape drawing and scaling.
    
    Parameters:
    - img_size (int): Size of the generated image before scaling.
    - num_pixels (int): Final size of the image after scaling.
    - color_palette (List[Tuple[int, int, int]]): List of available colors.
    
    Returns:
    - np.ndarray: Generated image.
    - List[dict]: Metadata for shapes in the image.
    """
    rng = np.random.default_rng()
    
    img = np.full((img_size,img_size,3),255,dtype=np.uint8)
    num_colours = len(color_palette)

    # Edge case protection, definitely need to figure out a better solution for min and max size
    min_size = img_size / 20
    max_size = img_size / 8

    num_shapes = rng.integers(3,max_shapes,endpoint=True)

    def rand_size(msize, xsize, metadata):
        placed = False
        for i in range(100): # 100 attempts to place the image, could be a failure point
            # some more edge case protection
            size = rng.integers(msize,xsize)
            center_x = rng.integers(2*size+1,(img_size-2*size)+1)
            center_y = rng.integers(2*size+1,(img_size-2*size)+1)
            center = (center_x,center_y)

            if not check_overlap(center, size, metadata):
                placed = True
                break

        if not placed:
            center, size = None, None
        
        return center, size
        
    def draw_shape(choice,img,center,size,colour):
        match choice:
            case 0:
                shape = 'square'
                area = draw_square(img,center,size,colour)
            case 1:
                shape = 'circle'
                area = draw_circle(img,center,size,colour)
            case 2:
                shape = 'triangle'
                area = draw_triangle(img,center,size,colour)

        return {'size':size,'center':center,'area':area,'shape':shape}
    
    metadata = []
    color_palette_copy = color_palette.copy()
    # Initial run to meet 1 square, 1 circle, 1 triangle requirement
    for shape in range(3):
        center, size = rand_size(min_size, max_size, metadata)
        
        # Forces these 3 to be different colours so we get some variety
        if len(color_palette_copy) > 0:
            rand_idx = rng.integers(0,len(color_palette_copy))
            colour = color_palette_copy.pop(rand_idx)

        if size is not None:
            metadata.append(draw_shape(shape,img,center,size,colour))
    
    for i in range(num_shapes-3):
        rand_c_idx = rng.integers(0,len(color_palette))
        colour = color_palette[rand_c_idx]

        rand_s_idx = rng.integers(0,3)
        center, size = rand_size(min_size, max_size, metadata)
        if size is not None:
            metadata.append(draw_shape(rand_s_idx,img,center,size,colour))
        
    # Allowed to use cv2 resize earlier, so will again here. Not sure the purpose of the scaling
    img = scale_image(img, num_pixels)
    
    return img, metadata

def assign_label(shapes_metadata: List[dict], fraction: float = 0.5) -> int:
    """
    Assigns a label based on a fractional comparison of the largest circle and square areas.
    
    Returns:
    - int: 0 if the largest square area is significantly larger than the largest circle area,
           2 if the largest circle area is significantly larger than the largest square area,
           1 otherwise.
    """
    # multiplying by fraction would make it smaller? I added 1 here else it doesn't make sense
    #if fraction < 1.0:
    fraction += 1.0

    biggest_circle = None
    biggest_square = None
    for shape_dict in shapes_metadata:  
        if shape_dict['shape'] == 'square':
            if not biggest_square or shape_dict['area'] > biggest_square['area']:
                biggest_square = shape_dict
        elif shape_dict['shape'] == 'circle':
            if not biggest_circle or shape_dict['area'] > biggest_circle['area']:
                biggest_circle = shape_dict

    if biggest_circle and biggest_square:
        if biggest_circle['area'] > biggest_square['area'] * fraction:
            return 2
        elif biggest_square['area'] > biggest_circle['area'] * fraction:
            return 0

    return 1


def generate_dataset(num_samples: int, img_size: int, num_pixels: int, num_colors: int, fraction: float = 2.0, max_shapes: int = 10) -> Tuple[np.ndarray, np.ndarray]:
    """
    Generates a dataset of images and labels in memory.
    
    Returns:
    - tuple[np.ndarray, np.ndarray]: Images (X) and labels (y).
    """
    colour_palette = generate_color_palette(num_colors)
    images = np.zeros((num_samples,num_pixels,num_pixels,3), dtype=np.uint8)
    labels = np.zeros(num_samples, dtype=int)
    
    for i in range(num_samples):
        img, metadata = generate_image(img_size, num_pixels, colour_palette, max_shapes)
        label = assign_label(metadata, fraction)

        images[i] = img
        labels[i] = label
    
    return (images, labels)
        

def display_sample_images(X: np.ndarray, y: np.ndarray, num_samples: int = 5) -> None:
    """
    Displays randomly sampled images from the dataset with labels.
    """
    rng = np.random.default_rng()

    # Finding num_samples unique and random images for the number of labels in y
    unique_labels = np.unique(y)
    num_labels = len(unique_labels)
    labels = {}

    indicies = np.arange(0,len(y))    
    for label in unique_labels:
        matches = indicies[label==y]

        if len(matches) >= num_samples:
            rand_idx = rng.choice(matches,num_samples,False)
            labels[label] = X[rand_idx]
        elif len(matches) > 0:
            rand_idx = matches
            labels[label] = X[rand_idx]
        else:
            labels[label] = []

    num_pixels = X[0].shape[0]
    figsize = (num_samples*4,8)
    f, axarr = plt.subplots(num_labels,num_samples,squeeze=False,figsize=figsize)
    for i in range(num_labels*num_samples):
        label_idx = i//num_samples
        label = unique_labels[label_idx]
        sample = i%num_samples
        if len(labels[label]) > sample:
            img = labels[label][sample]
            ax = axarr[label_idx,sample]
            ax.title.set_text(f"Label: {label}")
            ax.imshow(img)
            ax.axis('off')
    
save_history()
````

**ROC output convention:**
For `true_positive_rate` and `false_positive_rate`, sort `preds` in descending order and return cumulative rates after each sorted instance is included as positive. The returned arrays should therefore have the same length as `preds`. For `compute_tpr_fpr_range`, return `tpr_low`, `tpr_high`, `tpr_mid`, and `fpr` arrays with matching lengths, where the TPR summaries are quantiles across the supplied score lists at each FPR/threshold position.


````python
def filter_dataset(X: np.ndarray, y: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    """
    Filters the dataset to only include samples with labels 0 and 2,
    remaps label 2 to 1, and then balances the dataset so that both classes
    are represented equally.
    
    Parameters:
    - X (np.ndarray): Input images.
    - y (np.ndarray): Corresponding labels.
    
    Returns:
    - Tuple[np.ndarray, np.ndarray]: Balanced images and remapped labels.
    """
    rng = np.random.default_rng()
    
    indicies = np.arange(0,len(X))
    idx0 = indicies[y==0]
    idx1 = indicies[y==1]
    idx2 = indicies[y==2]

    idx1 = np.concatenate(idx1, idx2)
    smallest_array = len(idx0)
    if len(idx1) < smallest_array:
        smallest_array = len(idx1)

    rng.shuffle(idx0)
    rng.shuffle(idx1)
    idx0 = idx0[:smallest_array]
    idx1 = idx1[:smallest_array]
    idx_combined = np.concatenate(idx0, idx1)

    label_arr = y[idx_combined]
    label_arr[label_arr == 2] = 1 
    img_arr = X[idx_combined]

    return label_arr, img_arr

def true_positive_rate(preds: np.ndarray, targets: np.ndarray) -> np.ndarray:
    """Computes the True Positive Rate (TPR) for given predictions and targets."""
    target_values = np.unique(targets)
    if len(target_values) == 2:
        sort_tv = np.sort(target_values)
        negative = sort_tv[0]
        positive = sort_tv[1]
    else:
        negative = 0.0
        positive = 1.0
        
    sorted_preds = np.sort(preds, descending=True)
    sorted_targets = np.sort(targets, descending=True)
    true_positive = np.sum(np.logical_and(sorted_preds == positive, targets == positive))
    total_positives = preds[preds == positive]

    return true_positives / total_positives

def false_positive_rate(preds: np.ndarray, targets: np.ndarray) -> np.ndarray:
    """Computes the False Positive Rate (FPR) for given predictions and targets."""
    target_values = np.unique(targets)
    if len(target_values) == 2:
        sort_tv = np.sort(target_values)
        negative = sort_tv[0]
        positive = sort_tv[1]
    else:
        negative = 0.0
        positive = 1.0
        
    sorted_preds = np.sort(preds, descending=True)
    sorted_targets = np.sort(targets, descending=True)
    false_positive = np.sum(np.logical_and(sorted_preds == positive, targets == negative))
    total_negatives = preds[preds == negative]

    return false_positives / total_negatives

def compute_auc(TPR: np.ndarray, FPR: np.ndarray) -> float:
    """Computes the Area Under the Curve (AUC) for given TPR and FPR values."""
    return np.trapz(TPR, FPR)

def compute_tpr_fpr_range(
    scores_list: List[np.ndarray], 
    y_test: np.ndarray, 
    false_positive_rate_func: Callable[[np.ndarray, np.ndarray], np.ndarray], 
    true_positive_rate_func: Callable[[np.ndarray, np.ndarray], np.ndarray], 
    low_quantile: float, 
    high_quantile: float
) -> Tuple[List[float], List[float], List[float], List[float]]:
    """
    Computes the true positive rate (TPR) and false positive rate (FPR) ranges
    for multiple score distributions.

    Returns:
        - tpr_low: TPR at the `low_quantile`
        - tpr_high: TPR at the `high_quantile`
        - tpr_mid: Median TPR
        - fpr: List of false positive rates
    """
    fpr = []
    tpr = []
    for prediction in scores_list:
        fpr.append(false_positive_rate_func(prediction, y_test))
        tpr.append(true_positive_rate_func(prediction, y_test))

    fpr = np.sort(fpr)
    tpr = np.sort(tpr)
    tpr_len = len(tpr)
    tpr_mid = tpr[tpr_len//2]
    tpr_low = tpr[int(tpr_len*low_quantile)]
    tpr_high = tpr[int(tpr_len*high_quantile)]
    
    return tpr_low, tpr_high, tpr_mid, fpr
                               

def plot_roc(
    tpr_low: List[float], 
    tpr_high: List[float], 
    tpr_mid: List[float], 
    fpr: List[float], 
    compute_auc_func: Callable[[List[float], List[float]], float]
) -> None:
    """
    Plots the ROC curve with confidence intervals.

    Parameters:
        - tpr_low: Lower bound TPR values
        - tpr_high: Upper bound TPR values
        - tpr_mid: Median TPR values
        - fpr: False positive rate values
        - compute_auc_func: Function to compute AUC score
    """
    area = compute_auc_func(tpr_mid, fpr)

    # colour-blind friendly colours from https://gist.github.com/thriveth/8560036
    cb_cycle = ['#377eb8', '#ff7f00', '#4daf4a', '#f781bf', '#a65628', '#984ea3', '#999999', '#e41a1c', '#dede00']
    
    plt.figure()
    plt.title(f"AUR ROC: {area}")
    plt.plot([0,0.5,1],[0,0.5,1],label="Random Classifier",linestyle='dashed',alpha=0.8)
    plt.fill_between(fpr, tpr_high, tpr_low, alpha=0.5, color=cb_cycle[0])
    plt.plot(fpr, tpr_mid, label="Median TPR", color=cb_cycle[2], marker='o')
    plt.grid()
    plt.legend()
    plt.show()

save_history()
````

### **Weight Initialization Strategy**  

#### **Why Proper Initialization is Important?**  
Proper initialization of filter weights is **crucial** for stable training and efficient learning in deep networks. Poor initialization can lead to issues such as:  
- **Vanishing gradients** (if initialized too small).  
- **Exploding gradients** (if initialized too large).  
- **Slow convergence** if initialization does not match the expected scale of activations.  


#### **Initialization Strategies**

A common strategy is to initialize weights using a **scaled normal distribution** based on the number of input connections per neuron. The two widely used methods are:

- **Xavier (Glorot) Initialization**: Suitable for activations with **sigmoid** or **tanh** functions.  
  - Variance:  
    $$
    \frac{1}{\text{fan}_{\text{in}} + \text{fan}_{\text{out}}}
    $$
  
- **He Initialization**: Suitable for **ReLU-based** networks, where positive activations dominate.  
  - Variance:  
    $$
    \frac{2}{\text{fan}_{\text{in}}}
    $$
    (This helps prevent the outputs from shrinking.)

**How you should implement it in `ConvLayer`**  
You could use the **He initialization** since convolutional layers precede ReLU activation functions. The standard deviation can be scaled as follows:  

$$
\text{scale} = \sqrt{\frac{2}{\text{kernel\_size} \times \text{kernel\_size} \times \text{input\_channels}}}
$$

Each filter weight can be initialized using a normal distribution centered at `0` with this computed standard deviation:

```python
scale = np.sqrt(2.0 / (kernel_size * kernel_size * input_channels))
self.filters = np.random.randn(num_filters, input_channels, kernel_size, kernel_size) * scale
```

This would ensures that:  
- The weights start with an appropriate scale for stable learning.  
- Gradient magnitudes are balanced across layers, preventing vanishing/exploding gradients.  
- Each filter learns meaningful spatial features efficiently.  

By using this initialization strategy, the convolutional layer can **learn robust feature representations** from images while ensuring stable and efficient training.

### **Description of `dL_dout` in Backpropagation**  

`dL_dout` represents the **gradient of the loss function with respect to the output of the final layer** in the network. It is a key component in backpropagation, allowing the model to update its parameters by propagating the error signal backward through the layers.  

#### **How `dL_dout` is Computed**
- During training, the network computes predictions (`probs`), which are probability distributions obtained via the **softmax function**.
- The loss is typically computed using **categorical cross-entropy**, defined as:  
  $$
  L = - \sum_{i} y_i \log(\hat{y_i})
  $$
  where $ y $ is the true label (one-hot encoded) and $ \hat{y} $ is the predicted probability.
- The gradient of the loss with respect to the network’s output (logits) is:  
  $$
  dL_dout = \hat{y} - y
  $$
  This is derived from the fact that the derivative of cross-entropy loss combined with softmax simplifies to this expression.

#### **Shape of `dL_dout`**
- The output layer contains `num_classes` neurons, so `dL_dout` has shape **`(num_classes, 1)`**.
- Each element in `dL_dout` corresponds to how much the loss would change if the respective logit were adjusted.
- This gradient is then passed backward through the **fully connected layer**, allowing weight updates via gradient descent.

Thus, `dL_dout` serves as the starting point for backpropagation, determining how adjustments should be made to improve the network’s predictions.
